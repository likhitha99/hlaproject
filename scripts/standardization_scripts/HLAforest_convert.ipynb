```python
import pandas as pd
import glob
import numpy as np
```

```python
loc = '/content/gdrive/Shared drives/HLA/results/hlaforest/d3/'
df = pd.DataFrame()
```

```python
with open(loc+'/accession/d1_list.txt') as f:
  accessionnumbers = f.readlines()
```

```python
ERR,A1,A2,B1,B2,C1,C2 = list(), list(), list(), list(), list(), list(), list()
```

```python
for number in accessionnumbers: 
  print (number)
  # creating the path to the sample
  formatted_num = number.rstrip('\n')
  path = f"{loc}/{formatted_num}/haplotypes.txt"

  try:
    import csv
    limit = 0
    sample = list()
    with open(path) as bookfile:
      for line in bookfile:
        limit = limit +1
        if limit < 7:
          fields= line.split("\t")
          sample.append(fields[2][7:])
      #print(sample[1])
      ERR.append(number)
      A1.append(sample[0])
      A2.append(sample[1])
      B1.append(sample[2])
      B2.append(sample[3])
      C1.append(sample[4])
      C2.append(sample[5])

  except FileNotFoundError:
    pass
  except IndexError:
    pass
```

```python
for i in range(len(ERR)):
  ERR[i] = ERR[i].strip()
converted_list1=[]
for i in A1:
  converted_list1.append(i.strip())
converted_list2=[]
for i in A2:
  converted_list2.append(i.strip())
converted_list3=[]
for i in B1:
  converted_list3.append(i.strip())
converted_list4=[]
for i in B2:
  converted_list4.append(i.strip())
converted_list5=[]
for i in C1:
  converted_list5.append(i.strip())
converted_list6=[]
for i in C2:
  converted_list6.append(i.strip())
```

```python
dfERR = pd.DataFrame(ERR)
dfA1 = pd.DataFrame(converted_list1)
dfA2 = pd.DataFrame(converted_list2)
dfB1 = pd.DataFrame(converted_list3)
dfB2 = pd.DataFrame(converted_list4)
dfC1 = pd.DataFrame(converted_list5)
dfC2 = pd.DataFrame(converted_list6)
```

```python
df=pd.DataFrame(np.column_stack([ERR,converted_list1,converted_list2,converted_list3,converted_list4,converted_list5,converted_list6]), 
                               columns=['ERR','A', 'A.1','B','B.1','C','C.1'])
```

```python
for col in df.columns[df.columns!='ERR']:
    if '.' in col:
        allele = col.split('.')[0]
        for i in range(len(df[col])):
              df[col][i]=allele+'*'+ df[col][i]
    else: 
        for i in range(len(df[col])):
              df[col][i]=col+'*'+ df[col][i] 
```
