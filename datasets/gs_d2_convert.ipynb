```python
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv("2_gs_edited.csv")
df
```

```python
df.dropna(how='all', axis=1, inplace=True)
df
```

```python
for col in df.columns[df.columns!='Run']:
    if '.' in col:
        allele = col.split('.')[0]
        for i in range(len(df[col])):
            if '/' in df[col][i]:
                sample_list=df[col][i].split('/')
                for j in range(len(sample_list)):
                    value=allele+'*' + sample_list[j]
                    sample_list[j] = value
                string='/'.join(sample_list)
                df[col][i]=string
            else:
                df[col][i]=allele+'*'+ df[col][i]
    else: 
        for i in range(len(df[col])):
            if '/' in df[col][i]:
                sample_list=df[col][i].split('/')
                for j in range(len(sample_list)):
                    value=col+'*' + sample_list[j]
                    sample_list[j] = value
                string='/'.join(sample_list)
                df[col][i]=string
            else:
                df[col][i]=col+'*'+ df[col][i] 
```
