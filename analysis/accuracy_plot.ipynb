```python
df = pd.read_csv("/content/gdrive/MyDrive/test/accuracy_edited.csv")
df.head()
```

```python
df2 = pd.read_csv("/content/gdrive/MyDrive/test/accuracy.csv")
df2
```

```python
df.plot.bar(x='tool')
```

```python
sns.set_context("notebook")
ax = sns.barplot(x="tools", y="accuracy", data=df,ci="sd",hue="digit_type")
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy of HLA Callers Separated by Digit Resolution')
ax.set_xlabel('Tools')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
sns.despine()
```
