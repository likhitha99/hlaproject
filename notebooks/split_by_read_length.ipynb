```python
import tarfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sys import ps2
```

```python
arcas_d2=pd.read_csv("../results/read_length/arcasHLA_d2.csv",index_col="subject")
arcas_d3_d4=pd.read_csv("../results/read_length/arcasHLA_d3_4.csv",index_col="subject")
rna2hla_d2=pd.read_csv("../results/read_length/rna2hla_read_length_d2.csv",index_col="run")
rna2hla_d3_d4=pd.read_csv("../results/read_length/rna2hla_read_length.csv",index_col="Run")
phlat_d2=pd.read_csv("../results/read_length/phlat_d2_1.csv",index_col="Run")
phlat_d3_d4=pd.read_csv("../results/read_length/phlat_d3_d4.csv",index_col="Run")
```

```python
arcas=pd.concat([arcas_d2,arcas_d3_d4])
arcas=arcas.transpose()

rna2hla=pd.concat([rna2hla_d2,rna2hla_d3_d4])
rna2hla=rna2hla.transpose()

phlat=pd.concat([phlat_d2,phlat_d3_d4])
phlat=phlat.transpose()
```

```python
def split_csv(df,tool):    
    for i in range(36,127,5): #end at 127 so it includes 126

        csv_i = df.filter(regex="_"+str(i),axis=1)
        csv_i.columns = [name.split('_')[0] for name in csv_i.columns]
        csv_final = csv_i.transpose()
        #print(csv_final)
        csv_final.to_csv(f'../results/read_length/standard/{tool}_{str(i)}.csv')
```

```python
split_csv(arcas,'arcas')
split_csv(rna2hla,'rna2hla')
split_csv(phlat,'phlat')
```
