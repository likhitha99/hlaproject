```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

root = '..'
```

```python
# runtime plot
data = pd.read_csv(root+'/results/cpu_ram/runtime_data - cpu.csv')
df = data.melt()
sns.set_context("paper")
sns.color_palette("pastel")
sns.set(rc={'figure.figsize':(10,10)})
sns.set_style("white")

my_order = ["Optitype ","PHLAT","HLAforest","arcasHLA","rna2hla","seq2HLA","HLA-HD","HLAvbseq"]

ax = sns.boxplot(x="variable",y="value", data=df[12:],order=my_order)
ax.set_ylabel("CPU Time (s)")
ax.set_xlabel("Callers")
ax.set_title("CPU Time (s)") 
sns.despine()
plt.savefig(root+'/Figures/cpu.png',bbox_inches ="tight",dpi=300)
```

```python
# CPU time plot

gs = pd.read_csv(root+'/results/cpu_ram/runtime_data - ram.csv')
df = gs.melt()
sns.set_context("paper")
sns.color_palette("pastel")
sns.set(rc={'figure.figsize':(10,10)})
sns.set_style("white")

my_order = ["Optitype ","PHLAT","HLAforest","arcasHLA","rna2hla","seq2HLA","HLA-HD","HLAvbseq"]

ax = sns.boxplot(x="variable",y="value", data=df[12:],order=my_order)
ax.set_ylabel("RAM (bytes)")
ax.set_xlabel("Callers")
ax.set_title("RAM (bytes)") 
ax.set_yscale('log')
sns.despine()
plt.savefig(root+'/Figures/ram.png',bbox_inches ="tight",dpi=300)
```
