```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pylab as P

root = '..'
```

```python
# runtime plot
data = pd.read_csv(root+'/results/cpu_ram/runtime_data - cpu.csv')
df = data.melt()
sns.set_context("paper")
sns.color_palette("pastel")
sns.set(rc={'figure.figsize':(10,10)})
sns.set_style("white")

my_order = ["Optitype ","HLA-HD","PHLAT","HLAforest","arcasHLA","HLAMiner","rna2hla","seq2HLA","HLAvbseq"]

ax = sns.boxplot(x="variable",y="value", data=df[12:],order=my_order,showfliers=False)
ax = sns.stripplot(x="variable", y="value", data=df[12:],order=my_order,jitter=True,s=5,color='black',edgecolor='black')

ax.set_ylabel("CPU Time (s)")
ax.set_xlabel("Callers")
ax.set_title("CPU Time (s)") 
ax.set_yscale('log')
sns.despine()
plt.savefig(root+'/Figures/cpu.png',bbox_inches ="tight",dpi=300)
```

```python
# ram

gs = pd.read_csv(root+'/results/cpu_ram/runtime_data - ram.csv')
df = gs.melt()
sns.set_context("paper")
sns.color_palette("pastel")
sns.set(rc={'figure.figsize':(10,10)})
sns.set_style("white")

my_order = ["Optitype ","HLA-HD","PHLAT","HLAforest","arcasHLA","HLAMiner","rna2hla","seq2HLA","HLAvbseq"]

ax1 = sns.boxplot(x="variable",y="value", data=df[12:],order=my_order,showfliers=False)
ax1 = sns.stripplot(x="variable", y="value", data=df[12:],order=my_order,jitter=True,s=5,color='black',edgecolor='black')

ax1.set_ylabel("RAM (bytes)")
ax1.set_xlabel("Callers")
ax1.set_title("RAM (bytes)") 
ax1.set_yscale('log')
sns.despine()

plt.savefig(root+'/Figures/ram.png',bbox_inches ="tight",dpi=300)
```
