# CPUTime (CPU Utilized) = NCPUS * Elapsed (Wall-clock time).



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


from google.colab import drive
drive.mount('/content/gdrive')

root = '/content/gdrive/MyDrive/HLA_analysis/'
```

```python
# runtime plot
gs = pd.read_csv(root+'MASTER RESULTS DIRECTORY/runtime/ram.csv')
df = gs.melt()
sns.set_context("paper")
sns.color_palette("pastel")
sns.set(rc={'figure.figsize':(10,10)})
sns.set_style("white")
ax = sns.boxplot(x="variable",y="value", data=df[14:])
ax.set_ylabel("CPU Time (s)")
ax.set_xlabel("Callers")
ax.set_title("Dataset 1 CPU Time (s)") 
sns.despine()
plt.savefig(root+'/Figures/d1_ram.png',bbox_inches ="tight",dpi=300)
```
