```python
import csv 

with open('../read_length/standard/hlahd_76.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)

print(data)
```

```python
new_data = [['ERR', 'A', 'A.1', 'B', 'B.1', 'C', 'C.1', 'DRB1', 'DRB1.1', 'DQB1', 'DQB1.1']]
for line in data:
    l = [x for x in line if x.startswith('A*') or x.startswith('B*') or x.startswith('C*') or x.startswith('DQB1') or x.startswith('DRB1')]
    l.insert(0,line[0])
    try:
        if len(l) != 11:
            if (l[1][0] != 'A'):
                l.insert(1,'NA')
            if (l[2][0] != 'A'):
                l.insert(2,'NA')
            if (l[3][0] != 'B'):
                l.insert(3,'NA')
            if (l[4][0] != 'B'):
                l.insert(4,'NA')
            if (l[5][0] != 'C'):
                l.insert(5,'NA')
            if (l[6][0] != 'C'):
                l.insert(6,'NA')
            if (l[7][0:2] != 'DR'):
                l.insert(7,'NA')
            if (l[8][0:2] != 'DR'):
                l.insert(8,'NA')
            if (l[9][0:2] != 'DQ'):
                l.insert(7,'NA')
            if (l[10][0:2] != 'DQ'):
                l.insert(8,'NA')

    except:
        while len(l) != 11:
                l.insert(10,'NA')
        
    new_data.append(l)
print(new_data)
```

```python
import pandas as pd
df = pd.DataFrame(new_data)
df
```

```python
df.to_csv('../read_length/standard/hlahd_76.csv',index=False)
```
